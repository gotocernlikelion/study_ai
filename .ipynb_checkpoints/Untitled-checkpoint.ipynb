```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
my_data = []
for i in range(1000):
    x = np.random.normal(0,1)
    y = 0.1* x+ np.random.normal(0,0.05) # y = 0.1*x + 노이즈 
    my_data.append([x,y]) #
    #print(x)
#plt.plot(my_data)
```

```python
x_data = [ v[0] for v in my_data]
y_data = [ v[1] for v in my_data]
```

```python
# 랜덤으로 만든 x값에 0.1곱하고 노이즈 더해서 만든 데이터.
plt.figure(figsize=(7,6))
plt.plot(x_data, y_data,'*')
plt.show() 
```

```python
W = tf.Variable(tf.random_uniform([1],-1,1)) #1차원, -1부터 1 사이의 값. 
```

```python
b = tf.Variable(tf.zeros(1)) # b = 0
```

```python
hypothesis = W * x_data + b # y = Wx+b
cost = tf.reduce_mean(tf.square(hypothesis-y_data)) # 비용(MSE) = 평균(제곱(예측값 - 정답))
```

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.2)
train = optimizer.minimize(cost)
```

```python
sess = tf.Session() #session making
sess.run(tf.global_variables_initializer()) #변수 초기화 (반드시 필요해)
for step in range(10):
    print("step=",step, "Weight=", sess.run(W), "cost=", sess.run(cost))
    sess.run(train) #한번 돌때마다 1번씩 train이 돈다. train이 돌면 optimizer가 돌고, learning rate에 맞는 속도로 조금씩 W와 b를 조절해서 cost가 줄어든다. 
    plt.plot(x_data, y_data, '^')
    plt.plot(x_data, sess.run(W)*x_data+sess.run(b))
    #plt.show()
plt.show()
sess.close()
```

```python
x = np.linspace(1,10,1000)
y = 2*x+1 # y = W*x+b
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(1,activation='linear', input_shape=(1,) )) #tuple이니까 1 써준다. 
model.compile(loss='mse',optimizer='sgd')
```

```python
model.fit(x,y,epochs=20) #20번 학습하라. 
```
