```python
df = pd.read_excel("datasets/titanic.xls")

tmp = []
for i in df['sex']:
    if i == 'female':
        tmp.append(1)
    else:
        tmp.append(0)     
        
df['sex'] = tmp
df
```

```python
df = df[
    df['age'].notnull()
]
""" missing date 결측치를 버리고 남은 값만 모으기 """

df.info()
```

```python
df[ df['fare'].isnull() ]
""" True 인 값만 나온다 """
```

```python
df['fare'].isnull()
""" 항목별 True False여부만 나온다 """
```

```python
df = df [ df['fare'].notnull() ]
df
```

```python
df.info(
)
```

```python
#        pclass, sex, age, slisp, parch, fare
# index : 0       3    4    5      6      8

x_data = df.values[: , [0,3,4,5,6,8]] # values는 숫자만 따옴
x_data

```

```python
y_data = df.values[:,[1]] #생존여부 확인
y_data
```

```python
import random
random.seed(69) # 시드 고정하면, 결과가 항상 같게 유지됨
random.randint(1,100)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    x_data, y_data, test_size=0.1, random_state=7)
# x_data와 y_data가 각각 x _train x_test와 y_train, y_test로 스플릿 됨. 
# seed를 고정시키면 재연성이 올라감
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
X_train[0]
```

```python
y_train[0]
```

```python
# keras import 

import keras 
```

```python
from keras.models import Sequential #Sequential은 뇌 혹은 건물을 의미
from keras.layers import Dense #Dense는 한 층을 의미
```

```python
model = Sequential()
model.add(Dense(128, input_shape = (6, ), activation = 'relu')) #128개 뉴런, 6개의 input 요소, 다음 층으로는 relu함수(1층)
model.add(Dense(128, activation = 'relu')) #(2층)
model.add(Dense(1, activation = 'sigmoid'))#마지막 층의 경우 1개의 뉴런 보유, sigmoid 함수 사용 (3층)
#모델 설계
```

```python
model.compile(loss = 'mse', optimizer = 'adam', metrics = ['accuracy'])
#모델 설정
```

```python
hist = model.fit(X_train, y_train, validation_data = (X_test, y_test), epochs = 500) #epochs: 학습 회수
#훈련
```

```python
plt.figure(figsize = (12, 8))
plt.plot(hist.history['loss']) #훈련 비용
plt.plot(hist.history['val_loss']) #검증 비용
plt.plot(hist.history['acc']) #훈련 정확도
plt.plot(hist.history['val_acc']) #검증 정확도
plt.legend(['loss', 'val_loss', 'acc', 'val_acc'])
plt.show()
```
