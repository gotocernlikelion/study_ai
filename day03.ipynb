```python
from sklearn.datasets import load_iris
```

```python
iris_dataset = load_iris()
```

```python
from sklearn.model_selection import train_test_split # 원본 데이터 X를 섞어서, 프로그래머가 지정한 비율대로 훈련데이터와 테스트 데이터를 나누어준다. 
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    iris_dataset['data'], iris_dataset['target'], random_state=0 #시드값 고정
)
X_train.shape, X_test.shape # 디폴트 값은 3:1 로 쪼갠다. 즉 75%가 훈련 데이터 
print("훈련데이터는 약 ",int(len(X_train)/len(iris_dataset['data'])*100),"퍼센트")
```

```python
iris_df = pd.DataFrame(X_train, columns = iris_dataset['feature_names']) #훈련데이터를 데이터프레임 변수에 담고, 컬럼 명을 달아서 보기 좋게 함. 
iris_df
```

```python
iris_dataset['feature_names'] 
```

```python
pd.plotting.scatter_matrix( # pandas의 기능으로 scatter matrix 그래프를 그리겠다
    iris_df, # 이 데이터로
    c=y_train, # 0,1,2를 RGB 색상으로 쓰겠다는 말
    figsize=(15,15), #1515 사이즈
    marker='o', #동그라미
    hist_kwds={'bins':100}, # 대각선 방향에 히스토 그램
    s = 60,# 동그라미 크기
    alpha=.8
    )
plt.show()
```

```python
#iris 데이터는 최근접 이웃 알고리즘 (K near neighborhood, K neighborsC)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
 # 1 단계 : 모델 = 알고리즘( 하이퍼 파라미터 ) 
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
# 2 단계 : 모델 훈련
knn.fit(X_train, y_train)
```

```python
# 테스트용 데이터 하나 만들어 보기 
X_new = np.array([[5,2.9,1,0.2]]) # sepal length (cm)	sepal width (cm)	petal length (cm)	petal width (cm) 순
# 원본이 2차원이니까 새로운 데이터도 2차원으로 넣는다. 
```

```python
# 훈련한 걸로 한 번 예측해봄
prediction = knn.predict(X_new)
prediction 
# 0으로 예측함. 0 : Setosa
```

```python
# 맞나? 확인해보기 
print("예측한 품종 : ", iris_dataset['target_names'][prediction]) # prediction 이 0 이니까, target names의 0번째인 Setosa
# 정답!
# iris_dataset['target_names'] => ['setosa' 'versicolor' 'virginica']
```

```python
# 3단계-1st : 모델 평가
y_pred = knn.predict(X_test)
y_pred #예측값
```

```python
#3단계-2 답이 맞았나 체크해보기 : 모델 평가
```

```python
y_pred == y_test #False 가 있다. 
```

```python
np.mean(y_pred==y_test) 
# False = 0 , True = 1 로 더해서 평균을 구한다. 
#0.9736842105263158
#97.3% 의 test 의 정확도
```

```python
# 점수 매기기
knn.score(X_test,y_test)
```

```python
# 요약 
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
knn.score(X_test,y_test)
```
